<h2><div align="center">Line Chart</div></h2>

```python
w = np.linspace(0, 20, 200)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(w, np.sin(w))       # Plot the sine of each w point
plt.grid()                   # Show Grid
plt.show()                   # Display the plot
# https://www.youtube.com/watch?v=UUy6_ElQXBY&list=PLtPIclEQf-3dJmAj3IsSRwRoLbX-n3J81
```

```python
w = np.linspace(0, 2*np.pi, 100)
plt.plot(w, np.sin(w), color='red')       # Plot the sine of each w point
plt.show()                   # Display the plot
```

<h2><div align="center"> Sub-Plots </div></h2>

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)            #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40 ])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

```python
df = sns.load_dataset("penguins")
sns.jointplot(data=df,x="flipper_length_mm",y="bill_length_mm",hue="species", height=8)
plt.show()
```

```python
sns.pairplot(data=df, hue="species")
```

```python
tips = sns.load_dataset("tips")
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```
